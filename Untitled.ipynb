```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.colors import LinearSegmentedColormap

# Set style
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

# Load your results data (hierarchical bucket format [[primary, secondary, tertiary], window_size])
raw_results = {
  "[[1, 10, 60], 300]": 5521.7892651912025,
  "[[1, 10, 60], 600]": 5035.948162225186,
  "[[1, 10, 60], 900]": 5041.1608151839755,
  "[[1, 10, 60], 1200]": 4806.926704739806,
  "[[1, 10, 60], 1500]": 4643.237064875075,
  "[[1, 10, 60], 1800]": 4429.69454549117,
  "[[1, 10, 60], 2100]": 4260.928117998865,
  "[[1, 10, 60], 2400]": 4134.783709707738,
  "[[1, 10, 60], 2700]": 3970.3615540017217,
  "[[1, 10, 60], 3000]": 3817.739721566362,
  "[[1, 10, 60], 3300]": 3733.3332373583225,
  "[[1, 10, 60], 3600]": 3598.945862498205,
  "[[5, 60, 600], 300]": 4695.334357396972,
  "[[5, 60, 600], 600]": 3966.081959592095,
  "[[5, 60, 600], 900]": 3407.0895980897744,
  "[[5, 60, 600], 1200]": 5587.245780801837,
  "[[5, 60, 600], 1500]": 5387.440286923847,
  "[[5, 60, 600], 1800]": 5437.363895665944,
  "[[5, 60, 600], 2100]": 5404.8466373076335,
  "[[5, 60, 600], 2400]": 5474.173259069549,
  "[[5, 60, 600], 2700]": 5412.92538452319,
  "[[5, 60, 600], 3000]": 5406.769220496333,
  "[[5, 60, 600], 3300]": 5473.225735303891,
  "[[5, 60, 600], 3600]": 5360.523726259588,
  "[[10, 60, 600], 300]": 5738.1299468396155,
  "[[10, 60, 600], 600]": 5240.399490388391,
  "[[10, 60, 600], 900]": 4723.551162056409,
  "[[10, 60, 600], 1200]": 4419.709584566214,
  "[[10, 60, 600], 1500]": 4077.025094609084,
  "[[10, 60, 600], 1800]": 3734.5932545585124,
  "[[10, 60, 600], 2100]": 3522.7241261439653,
  "[[10, 60, 600], 2400]": 5874.413640193362,
  "[[10, 60, 600], 2700]": 5822.831620932173,
  "[[10, 60, 600], 3000]": 5799.92289903193,
  "[[10, 60, 600], 3300]": 5772.765683948241,
  "[[10, 60, 600], 3600]": 5682.930998256676,
  "[[30, 300, 900], 300]": 6376.029431609532,
  "[[30, 300, 900], 600]": 6106.9638312202405,
  "[[30, 300, 900], 900]": 5874.078271198291,
  "[[30, 300, 900], 1200]": 5801.263344799713,
  "[[30, 300, 900], 1500]": 5612.5146927654,
  "[[30, 300, 900], 1800]": 5459.568577264391,
  "[[30, 300, 900], 2100]": 5412.261029008525,
  "[[30, 300, 900], 2400]": 5058.951627378237,
  "[[30, 300, 900], 2700]": 4970.689247219581,
  "[[30, 300, 900], 3000]": 4936.430162019479,
  "[[30, 300, 900], 3300]": 4830.4999587750735,
  "[[30, 300, 900], 3600]": 4634.443096737676
}

# Convert to DataFrame with hierarchical bucket information
data = []
for key, ips in raw_results.items():
    bucket_hierarchy, window_size = eval(key)
    primary, secondary, tertiary = bucket_hierarchy
    
    # Create a unique identifier for the bucket hierarchy
    hierarchy_key = f"{primary}s-{secondary}s-{tertiary}s"
    
    # Determine which bucket sizes would be used based on window size
    # This is an approximation - the actual usage depends on your implementation logic
    active_buckets = []
    if window_size <= 300:  # 5m
        active_buckets = [primary, secondary]
    elif window_size <= 1800:  # 30m
        active_buckets = [primary, secondary, tertiary]
    else:  # > 30m
        active_buckets = [secondary, tertiary]
    
    # Calculate the estimated bucket storage based on active bucket sizes
    # This is a simplified model - actual storage depends on implementation details
    storage_estimate = 0
    for bucket_size in active_buckets:
        # Estimate how many buckets of this size would be needed for the window
        buckets_needed = min(window_size // bucket_size, window_size // bucket_size + 1)
        storage_estimate += buckets_needed
    
    data.append({
        'hierarchy_key': hierarchy_key,
        'primary': primary,
        'secondary': secondary,
        'tertiary': tertiary,
        'window_size': window_size,
        'ips': ips,
        'active_buckets': active_buckets,
        'estimated_storage': storage_estimate,
        'storage_efficiency': ips / storage_estimate if storage_estimate > 0 else 0
    })

df = pd.DataFrame(data)

# Create a figure with 3 rows, 2 columns layout for better readability
fig = plt.figure(figsize=(20, 24))

# 1. Performance by Hierarchy and Window Size (Heat Map)
ax1 = plt.subplot(3, 2, 1)
pivot_df = df.pivot(index='hierarchy_key', columns='window_size', values='ips')
heatmap = sns.heatmap(pivot_df, annot=True, fmt='.0f', cmap='RdYlGn', ax=ax1, 
                     cbar_kws={'label': 'Operations per Second'})
ax1.set_title('Operations per Second by Bucket Hierarchy\n(Higher is Better)', fontsize=16, fontweight='bold')
ax1.set_ylabel('Bucket Hierarchy [primary-secondary-tertiary]', fontsize=14)
ax1.set_xlabel('Window Size (seconds)', fontsize=14)
ax1.tick_params(axis='both', which='major', labelsize=12)

# 2. Performance by Window Size (Line Plot)
ax2 = plt.subplot(3, 2, 2)
hierarchies = sorted(df['hierarchy_key'].unique())
colors = plt.cm.Set3(np.linspace(0, 1, len(hierarchies)))

for hierarchy, color in zip(hierarchies, colors):
    subset = df[df['hierarchy_key'] == hierarchy].sort_values('window_size')
    ax2.plot(subset['window_size'], subset['ips'], 
             marker='o', linewidth=3, markersize=10,
             label=hierarchy, color=color)

ax2.set_xlabel('Window Size (seconds)', fontsize=14)
ax2.set_ylabel('Operations per Second', fontsize=14)
ax2.set_title('Performance Scaling by Bucket Hierarchy', fontsize=16, fontweight='bold')
ax2.legend(fontsize=12, loc='best')
ax2.grid(True, alpha=0.3)
ax2.tick_params(axis='both', which='major', labelsize=12)

# 3. Performance vs Estimated Storage
ax3 = plt.subplot(3, 2, 3)
for hierarchy, color in zip(hierarchies, colors):
    subset = df[df['hierarchy_key'] == hierarchy].sort_values('estimated_storage')
    ax3.plot(subset['estimated_storage'], subset['ips'], 
             marker='o', linewidth=3, markersize=10,
             label=hierarchy, color=color)

ax3.set_xlabel('Estimated Storage (# of entries in Redis)', fontsize=14)
ax3.set_ylabel('Operations per Second', fontsize=14)
ax3.set_title('Performance vs Storage Usage\n(Key Resource Trade-off)', fontsize=16, fontweight='bold')
ax3.legend(fontsize=12, loc='best')
ax3.grid(True, alpha=0.3)
ax3.tick_params(axis='both', which='major', labelsize=12)

# 4. Storage Efficiency (Operations per Estimated Storage)
ax4 = plt.subplot(3, 2, 4)
for hierarchy, color in zip(hierarchies, colors):
    subset = df[df['hierarchy_key'] == hierarchy].sort_values('window_size')
    ax4.plot(subset['window_size'], subset['storage_efficiency'], 
             marker='o', linewidth=3, markersize=10,
             label=hierarchy, color=color)

ax4.set_xlabel('Window Size (seconds)', fontsize=14)
ax4.set_ylabel('Ops per Second / Storage Unit', fontsize=14)
ax4.set_title('Storage Efficiency by Window Size\n(Higher = More Efficient)', fontsize=16, fontweight='bold')
ax4.legend(fontsize=12, loc='best')
ax4.grid(True, alpha=0.3)
ax4.tick_params(axis='both', which='major', labelsize=12)

# 5. Box Plot of Performance by Hierarchy
ax5 = plt.subplot(3, 2, 5)
box_plot = sns.boxplot(data=df, x='hierarchy_key', y='ips', ax=ax5, palette='Set3')

# Add mean values on top of boxes
for i, hierarchy in enumerate(hierarchies):
    subset = df[df['hierarchy_key'] == hierarchy]
    mean_val = subset['ips'].mean()
    ax5.text(i, mean_val + 200, f'μ={mean_val:.0f}', ha='center', fontsize=11, fontweight='bold')

ax5.set_xlabel('Bucket Hierarchy [primary-secondary-tertiary]', fontsize=14)
ax5.set_ylabel('Operations per Second', fontsize=14)
ax5.set_title('Performance Distribution by Bucket Hierarchy\n(with means)', fontsize=16, fontweight='bold')
ax5.tick_params(axis='both', which='major', labelsize=12)

# 6. Performance Variation Analysis
ax6 = plt.subplot(3, 2, 6)
performance_ranges = []
for hierarchy in hierarchies:
    subset = df[df['hierarchy_key'] == hierarchy]
    min_perf = subset['ips'].min()
    max_perf = subset['ips'].max()
    performance_drop = ((max_perf - min_perf) / max_perf) * 100
    
    # Find longest streak of consistent performance
    subset_sorted = subset.sort_values('window_size')
    streak_length = 0
    max_streak = 0
    baseline = None
    for _, row in subset_sorted.iterrows():
        current = row['ips']
        if baseline is None:
            baseline = current
            streak_length = 1
        elif abs(current - baseline) / baseline < 0.1:  # Within 10% of baseline
            streak_length += 1
        else:
            baseline = current
            max_streak = max(max_streak, streak_length)
            streak_length = 1
    max_streak = max(max_streak, streak_length)
    
    performance_ranges.append({
        'hierarchy_key': hierarchy,
        'drop_percentage': performance_drop,
        'consistency_streak': max_streak
    })

drop_df = pd.DataFrame(performance_ranges)
bars = ax6.bar(drop_df['hierarchy_key'], drop_df['drop_percentage'], 
               color=['darkred' if x > 80 else 'red' if x > 50 else 'orange' if x > 20 else 'green' for x in drop_df['drop_percentage']],
               alpha=0.8)

# Add a second axis for consistency streak
ax6_twin = ax6.twinx()
ax6_twin.plot(drop_df['hierarchy_key'], drop_df['consistency_streak'], 'bo-', linewidth=2, markersize=8)
ax6_twin.set_ylabel('Consistency Streak (# of windows)', color='blue', fontsize=14)
ax6_twin.tick_params(axis='y', labelcolor='blue')

ax6.set_xlabel('Bucket Hierarchy [primary-secondary-tertiary]', fontsize=14)
ax6.set_ylabel('Performance Variation (%)', fontsize=14, color='black')
ax6.set_title('Performance Stability Analysis\n(Lower Variation + Higher Streak = Better)', fontsize=16, fontweight='bold')
ax6.tick_params(axis='both', which='major', labelsize=12)

# Add value labels on bars with better formatting
for bar, value in zip(bars, drop_df['drop_percentage']):
    height = bar.get_height()
    ax6.text(bar.get_x() + bar.get_width()/2., height + 1,
             f'{value:.1f}%', ha='center', va='bottom', fontsize=12, fontweight='bold')

# Add a horizontal line at 20% as a "danger zone" indicator
ax6.axhline(y=20, color='orange', linestyle='--', alpha=0.7, label='20% (Caution)')
ax6.axhline(y=50, color='red', linestyle='--', alpha=0.7, label='50% (Danger)')
ax6.legend(fontsize=11, loc='upper left')

plt.tight_layout(pad=3.0)
plt.show()

# Print Enhanced Summary Statistics for Hierarchical Buckets
print("=== HIERARCHICAL BUCKET PERFORMANCE ANALYSIS ===\n")

print("1. BEST OVERALL PERFORMANCE:")
best = df.loc[df['ips'].idxmax()]
print(f"   Hierarchy [{best['primary']}s, {best['secondary']}s, {best['tertiary']}s] with {best['window_size']}s window: {best['ips']:.0f} ops/sec")

print("\n2. WORST OVERALL PERFORMANCE:")
worst = df.loc[df['ips'].idxmin()]
print(f"   Hierarchy [{worst['primary']}s, {worst['secondary']}s, {worst['tertiary']}s] with {worst['window_size']}s window: {worst['ips']:.0f} ops/sec")
print(f"   Performance difference: {best['ips']/worst['ips']:.1f}x")

print("\n3. PERFORMANCE VARIABILITY BY HIERARCHY:")
for hierarchy in sorted(hierarchies):
    subset = df[df['hierarchy_key'] == hierarchy]
    primary = subset['primary'].iloc[0]
    secondary = subset['secondary'].iloc[0]
    tertiary = subset['tertiary'].iloc[0]
    min_ips = subset['ips'].min()
    max_ips = subset['ips'].max()
    mean_ips = subset['ips'].mean()
    variation = ((max_ips - min_ips) / max_ips) * 100
    print(f"   [{primary:2d}s, {secondary:3d}s, {tertiary:3d}s]: {variation:5.1f}% variation (min: {min_ips:4.0f}, max: {max_ips:4.0f}, avg: {mean_ips:4.0f})")

print("\n4. STORAGE EFFICIENCY (ops/sec per storage unit):")
efficiency_by_hierarchy = df.groupby('hierarchy_key')['storage_efficiency'].mean().sort_values(ascending=False)
for hierarchy, eff in efficiency_by_hierarchy.items():
    subset = df[df['hierarchy_key'] == hierarchy]
    primary = subset['primary'].iloc[0]
    secondary = subset['secondary'].iloc[0]
    tertiary = subset['tertiary'].iloc[0]
    print(f"   [{primary:2d}s, {secondary:3d}s, {tertiary:3d}s]: {eff:6.0f} ops/sec per storage unit")

print("\n5. WINDOW SIZE PERFORMANCE ANALYSIS:")
for window_size in [300, 1800, 3600]:  # 5m, 30m, 1h
    print(f"\n   {window_size}s window performance:")
    window_data = df[df['window_size'] == window_size].sort_values('ips', ascending=False)
    for _, row in window_data.iterrows():
        hierarchy = row['hierarchy_key']
        primary = row['primary']
        secondary = row['secondary']
        tertiary = row['tertiary']
        print(f"   [{primary:2d}s, {secondary:3d}s, {tertiary:3d}s]: {row['ips']:4.0f} ops/sec")

print("\n6. RECOMMENDATION ANALYSIS:")
# Find best hierarchy based on combined metrics
stability_scores = []
for hierarchy in hierarchies:
    subset = df[df['hierarchy_key'] == hierarchy]
    primary = subset['primary'].iloc[0]
    secondary = subset['secondary'].iloc[0]
    tertiary = subset['tertiary'].iloc[0]
    
    # Calculate key metrics
    variation = ((subset['ips'].max() - subset['ips'].min()) / subset['ips'].max()) * 100
    avg_perf = subset['ips'].mean()
    min_perf = subset['ips'].min()
    perf_score = avg_perf / df['ips'].max() * 100  # Percent of max possible performance
    
    # Higher is better for this combined score
    combined_score = (100 - variation) * 0.5 + perf_score * 0.5
    
    stability_scores.append({
        'hierarchy': hierarchy,
        'primary': primary,
        'secondary': secondary,
        'tertiary': tertiary,
        'variation': variation,
        'avg_perf': avg_perf,
        'min_perf': min_perf,
        'combined_score': combined_score
    })

stability_df = pd.DataFrame(stability_scores).sort_values('combined_score', ascending=False)
best_hierarchy = stability_df.iloc[0]
print(f"   Best Overall Hierarchy: [{best_hierarchy['primary']}s, {best_hierarchy['secondary']}s, {best_hierarchy['tertiary']}s]")
print(f"   - Performance Score: {best_hierarchy['combined_score']:.1f}%")
print(f"   - Performance Variation: {best_hierarchy['variation']:.1f}%")
print(f"   - Average Performance: {best_hierarchy['avg_perf']:.0f} ops/sec")
print(f"   - Minimum Performance: {best_hierarchy['min_perf']:.0f} ops/sec")

# Print out the second-best option for comparison
second_best = stability_df.iloc[1]
print(f"\n   Second-Best Hierarchy: [{second_best['primary']}s, {second_best['secondary']}s, {second_best['tertiary']}s]")
print(f"   - Performance Score: {second_best['combined_score']:.1f}%")
print(f"   - Performance Variation: {second_best['variation']:.1f}%")
print(f"   - Average Performance: {second_best['avg_perf']:.0f} ops/sec")
print(f"   - Minimum Performance: {second_best['min_perf']:.0f} ops/sec")

# Print specific window size recommendations
print("\n7. WINDOW-SPECIFIC RECOMMENDATIONS:")
for window_size in [300, 900, 1800, 3600]:
    window_data = df[df['window_size'] == window_size].sort_values('ips', ascending=False)
    best_for_window = window_data.iloc[0]
    primary = best_for_window['primary']
    secondary = best_for_window['secondary']
    tertiary = best_for_window['tertiary']
    print(f"   {window_size}s window: Use [{primary}s, {secondary}s, {tertiary}s] for {best_for_window['ips']:.0f} ops/sec")
```
